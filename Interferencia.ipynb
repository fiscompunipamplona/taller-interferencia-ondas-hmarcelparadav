# Representacin de las ondas separadas una distancia de 20 cm

```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin

pix = 500
size = 100. 
step = size / pix 
lda = 5.0 
k = 2.*pi / lda
zeta0 = 1.0 
Xi = float(input("Introduzca la coordenada de la onda en X en cm:")) 
Yi = float(input("Introduzca la coordenada de la onda en Y en cm:"))

x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
ondulacion = open("outdata.dat", "w")

for yy in range( len(x) ):
    for xx in range( len(y) ):
        ri = sqrt( (Xi-x[xx])*(Xi-x[xx]) + (Yi-y[yy])*(Yi-y[yy]) )
        r1 = sqrt( ((Xi+20)-x[xx])*((Xi+20)-x[xx]) + (Yi-y[yy])*(Yi-y[yy]) ) 
        zeta1=zeta0*sin(ri*k)
        zeta2=zeta0*sin(r1*k)
        ondulacion.write("%0.1f " % (zeta1))
        ondulacion.write("%0.1f " % (zeta2))
    ondulacion.write("\n")
    
ondulacion.close()

data = loadtxt("outdata.dat", float)
imshow(data,origin="lower",extent=[0,100,0,100])
xlabel("Trnasmisin de la onda en X cm ")
ylabel("Transmisin de la onda en Y cm")
imshow(data)
show()
```

# En el siguiente grfico se representa la interferencia de las dos ondas, teniendo en cuenta la separacin de 20cm

```python
from pylab import plot, imshow, show,xlabel,ylabel
from numpy import loadtxt, array, zeros, sqrt, pi,sin

pix = 500
size = 100. 
step = size / pix 
lda = 5.0 
k = 2.*pi / lda
zeta0 = 1.0 
print("Interferencia")

Xi = 50. 
Yi = 50. 
x = zeros(pix, float)
y = zeros(pix, float)

for i in range(pix):
    x[i] = step*i
    y[i] = step*i
    
ondulacion = open("outdata.dat", "w")

for yy in range( len(x) ):
    for xx in range( len(y) ):
        ri = sqrt((Xi-x[xx])*(Xi-x[xx]) + (Yi-y[yy])*(Yi-y[yy]) )
        r1 = sqrt(((Xi+20)-x[xx])*((Xi+20)-x[xx]) + (Yi-y[yy])*(Yi-y[yy]) )
        zeta1=zeta0*sin(ri*k)
        zeta2=zeta0*sin(r1*k)
        
        ondulacion.write("%0.1f " % (zeta1))
        ondulacion.write("%0.1f " % (zeta2))
        ondulacion.write("%0.1f " % (zeta1+zeta2))
    
    ondulacion.write("\n")
    
ondulacion.close()

data = loadtxt("outdata.dat", float)
imshow(data,origin="lower", extent=[0,100,0,100])
xlabel("Transmisin de la onda en X cm")
ylabel("Transmisin de la onda en Y cm")
show()
```
